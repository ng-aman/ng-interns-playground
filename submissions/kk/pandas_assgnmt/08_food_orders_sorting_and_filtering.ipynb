### Create Dataframe chipotle.tsv Assign it to a variable called chipo.

```python
chipo = pd.read_table('chipotle 1.tsv')
chipo.head()
```

```python
# Question 1. How many products in the dataset cost more than $10.00 using apply and lambda?
chipo[chipo['item_price'].apply(lambda x: float(x[1:])) > 10.00]
```

```python
# Question 2. What is the price of each item? Can you provide a data frame with only two columns: item_name and item_price?
chipo[['item_name', 'item_price']]
```

```python
# Question 3. How many different types of Chicken Bowl were ordered? Can you list them along with their choice_description and item_price?
chick_bowl = chipo[chipo['item_name']=='Chicken Bowl']
chick_bowl[['item_name','choice_description','item_price']]
```
